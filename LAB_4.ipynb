```python

import seaborn as sns
iris = sns.load_dataset('iris')

#20
min_value = iris[iris['species'] == 'virginica']['sepal_width'].min()

print(min_value)

#21
mean_petal_length = iris['petal_length'].mean()
print(mean_petal_length)
filtered_df = iris[iris['petal_length'] > mean_petal_length]
count = filtered_df.shape[0]
print(count)
#22
min_sepal=iris['sepal_length'].mean()
print("sepal mean "+str(min_sepal))
species= iris['species'].value_counts()
print(species)
unique_species = iris['species'].nunique()
print(unique_species)

#23
corr = iris.corr(numeric_only=True)

print(corr)
#24
#used ai to learn new built in functions
summary = iris.groupby('species').agg(['mean', 'std'])
print("24: "+str(summary))


#25
import seaborn as sns
titanic = sns.load_dataset('titanic')

survived = titanic[(titanic['pclass'] == 2) & (titanic['survived'] == 1)]

count1 = survived.shape[0]
print(count1)

#26
died = titanic[(titanic['pclass'] == 1) & (titanic['survived'] == 0)]
age_min = died['age'].min()
age_max = died['age'].max()
print(f"Age range: {age_min}-{age_max}")


#27
female_passengers = titanic[titanic['sex'] == 'female']
female_survived = female_passengers[female_passengers['survived'] == 1].shape[0]
total_females = female_passengers.shape[0]
percentage = round((female_survived / total_females) * 100)
print(f"Percentage of female passengers who survived: {percentage}%")

#28
missing_counts = titanic.isnull().sum()
print(missing_counts)
```
