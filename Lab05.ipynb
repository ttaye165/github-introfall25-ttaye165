```python
import pandas as pd

# Load dataset (adjust filename if needed)
df = pd.read_csv("spam.csv")

# Inspect columns to see what they look like
print("Original columns:", df.columns)

# Rename the key columns to something clearer
# In most versions of the dataset: v1 = label, v2 = sms
df = df.rename(columns={"v1": "label", "v2": "sms"})

# Keep only the relevant columns
df = df[['label', 'sms']]

# Ensure 'sms' is treated as string
df['sms'] = df['sms'].astype(str)

#26
spamonly = df[df['label'] == 'spam']
duplicates_in_spam = spamonly.duplicated(subset=['sms']).sum()
print("Number of duplicates in spamonly:", duplicates_in_spam)
#27
spamonly_nodup = spamonly.drop_duplicates(subset=['sms'])
total_spam_sms = len(spamonly_nodup)
print(total_spam_sms)

#28
df_nodup = df.drop_duplicates(subset=['sms'])
avg_length = round(df_nodup['sms'].apply(len).mean(), 2)
print(avg_length)

#29
df_nodup = df.drop_duplicates(subset=['sms'])
df_nodup.loc[:, 'sentences_count'] = df_nodup['sms'].str.split('.').str.len()
max_sentences = df_nodup['sentences_count'].max()
print(max_sentences)
#30
df_nodup['words_count'] = df_nodup['sms'].str.split().str.len()
min_words = df_nodup['words_count'].min()
print(min_words)
#31
# remove duplicates
df_nodup = df.drop_duplicates(subset=['sms']).copy()

# take first 5 messages
first_five = df_nodup.head(5)

# count capital letters in each message
first_five['capital_count'] = first_five['sms'].str.count(r'[A-Z]')

# highest number of capital letters
max_capitals = first_five['capital_count'].max()
print(max_capitals)
```
